```python
def remove_outliers_iqr(feature):
    Q1 = df[feature].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[feature].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1                       # Interquartile range

    ### Research about these indices for outlier determination 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the data to keep only non-outliers
    return df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
```

```python
# Randomly sample 10 records
sampled_df = df.sample(n=10,random_state=10)  # Set random_state for reproducibility
sampled_df['Income'].mean()
```

```python
#variables - parameters
sample_mean = sampled_df['Income'].mean()
sample_sd = sampled_df['Income'].std()

###### for known distribution it gets the population (deck nr4)
pop_sd = df['Income'].std()
pop_mean = df['Income'].mean()

n = len(sampled_df)
n
```

```python
## Standard Error for the sample (using the population) 
SE = pop_sd / np.sqrt(n)
```

```python

# Z-value for 97.5% confidence
z = stats.norm.ppf(0.975)  # two-tailed
```

```python
# Calculate the confidence interval using T-value
confidence_interval = (
    sample_mean - z * SE,
    sample_mean + z * SE
)
```

```python
    confidence_interval
```

```python

# Output the results
print("Standard Error", SE)
print(" Mean:", sample_mean)
print(" Standard Deviation:", pop_sd)
print("97.5% Confidence Interval:", confidence_interval)
```
